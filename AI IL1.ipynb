```python
nums = [5, 2, 8, 1, 9]
print("Original:", nums)

nums.append(4)
print("After append:", nums)

nums.insert(2, 7)
print("After insert:", nums)

nums.remove(8)
print("After remove:", nums)

nums.sort()
print("After sort:", nums)

del nums[1]
print("After delete:", nums)
```

```python
nums = (5, 2, 8, 1, 9)
print("Original:", nums)

# TUPLES ARE IMMUTABLE - CONVERT TUPLES TO LIST FIRST
lst = list(nums)

# NOW PERFORM OPERATIONS
lst.append(4)
nums = tuple(lst)
print("After append:", nums)

lst = list(nums)
lst.insert(2, 7)
nums = tuple(lst)
print("After insert:", nums)

lst = list(nums)
lst.remove(1)
nums = tuple(lst)
print("After remove:", nums)

lst = list(nums)
lst.sort()
nums = tuple(lst)
print("After sort:", nums)

lst = list(nums)
del lst[1]
nums = tuple(lst)
print("After delete:", nums)
```

```python
s = {5, 2, 8, 1, 9}
print("Original:", s)

s.add(4)
print("After add:", s)

s.update([10, 11])
print("After update:", s)

s.remove(8)
print("After remove:", s)

print("Sorted:", sorted(s))

s.clear()
print("After clear:", s)
```

```python
students = {"Alice": 85, "Bob": 90, "Charlie": 78}
print("Original:", students)

students["David"] = 88
print("After adding:", students)

students["Bob"] = 95
print("After updating:", students)

del students["Charlie"]
print("After deleting:", students)

print("Names:", list(students.keys()))
print("Marks:", list(students.values()))
```

```python
marks = int(input("Enter marks in range of 0-100): "))
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: F")
```

```python
def add(a, b = 5):
    return a + b

def greet(name="Guest"):
    return f"Hello, {name}!"

def power(base, exp = 2):
    return base ** exp

print(add(3))
print(add(3, 4))
print(greet())
print(greet("Zubair"))
print(power(3))
print(power(3, 3))
```

```python
def simple_interest(PrincipalAmount, Timeperiod, InterestRate = 5):
    return (PrincipalAmount * InterestRate * Timeperiod) / 100

print(simple_interest(1000, 2))
print(simple_interest(1000, 2, 10))
```
