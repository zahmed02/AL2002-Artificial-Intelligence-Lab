### 1. Write a function to calculate the sum of two numbers, where the second number has a default value of 5. Call the function with one and two arguments.
### 2. Write a function to greet a user with their name, where the name has a default value “Guest”.
### 3. Write a function to calculate the power of a number, where the exponent has a default value of 2 (i.e., square by default).

```python
def add(a, b = 5):
    return a + b

def greet(name = "Guest"):
    return f"Hello, {name}!"

def power(base, exp = 2):
    return base ** exp

print(add(3))
print(add(3, 4))
print(greet())
print(greet("Zubair"))
print(power(3))
print(power(3, 3))
```

## AI L1 T10
### Write a function to compute simple interest and return it, where the rate has a default value of 5%.

```python
def simpleinterest(PrincipalAmount, Timeperiod, InterestRate = 5):
    return (PrincipalAmount * InterestRate * Timeperiod) / 100

print(simpleinterest(1000, 2))
print(simpleinterest(1000, 2, 10))
```
